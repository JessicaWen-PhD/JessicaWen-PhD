# Read Gaussian Cube File
This notebook aims to read, parse, and visualise Gaussian Cube files generated by Environ.

```python
import ase
from ase.io.cube import read_cube_data
```

```python
filename = "velectrostatic.cube"
velectrostatic = read_cube_data(filename)
print(velectrostatic)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Unpack: first element is 3D array (nx, ny, nz), second is Atoms with cell
data, atoms = velectrostatic
nx, ny, nz = data.shape
cell = atoms.cell

# Average potential in each xy-layer (constant z): average over axes 0 and 1
v_avg_z = data.mean(axis=(0, 1))

# z coordinates: center of each layer along the cell's third axis (Å)
# Assumes grid is aligned with cell; origin at 0
dz = cell[2, 2] / nz
z_vals = (np.arange(nz) + 0.5) * dz

# Plot average potential vs z
fig, ax = plt.subplots()
ax.plot(z_vals, v_avg_z)
ax.set_xlabel("z (Å)")
ax.set_ylabel("Planar-averaged potential (eV)")
ax.set_title("Planar-Averaged Electrostatic Potential along z")
plt.tight_layout()
plt.show()
```
